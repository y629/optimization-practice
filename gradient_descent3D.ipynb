```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 適当な範囲内の関数の値と点列をプロットする関数
def show3D(X, Y, Z, x_pred, y_pred, z_pred):
    # フォントの設定
    plt.rcParams['font.size'] = 14
    plt.rcParams['font.family'] = 'Times New Roman'
    # グラフの入れ物を作成
    fig = plt.figure()
    ax1 = Axes3D(fig)
    # 軸のラベルを設定する
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('z')
     # データプロットする
    ax1.plot_wireframe(X, Y, Z, label='f(x, y)')
    ax1.scatter3D(x_pred, y_pred, z_pred, label='gd', color='red', s=10)
    # グラフを表示する
    plt.show()
    plt.close()
```

```python
# 関数とその勾配を定義
def f(x, y):
#     return x**2 - y**2
    return np.log(np.exp(1+x)+np.exp(1+y))

def nabla_f(x, y):
#     return 2 * np.array([x, y])
    return np.array([np.exp(1+x), np.exp(1+y)])/(np.exp(1+x)+np.exp(1+y))
```

```python
# 指定したstep数だけ最急勾配法を行う関数
def steepest(x0, l, step, f, nabla_f):
    x = np.ndarray((step, x0.shape[0]))
    x[0] = x0
    f_history = np.ndarray((step))
    f_history[0] = f(*x[0])
    for i in range(step):
        if i == 0:
            continue
        x[i] = x[i-1] - nabla_f(*x[i-1]) / l

        f_history[i] = f(*x[i])
    return x, f_history
```

```python
l, step = 10, 500
"""ケース1"""
# 初期点を設定してアルゴリズムを実行
x0 = np.array([10, 0])
x, f_history = steepest(x0, l, step, f, nabla_f)
# グラフを描画
x_pred = np.array([x[i][0] for i in range(step)])
y_pred = np.array([x[i][1] for i in range(step)])
z_pred = f(x_pred, y_pred)
x = np.arange(-10, 11, 2)
y = np.arange(-10, 11, 2)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
show3D(X, Y, Z, x_pred, y_pred, z_pred)
"""ケース2"""
# 初期点を設定してアルゴリズムを実行
x0 = np.array([-5, 10])
x, f_history = steepest(x0, l, step, f, nabla_f)
# グラフを描画
x_pred = np.array([x[i][0] for i in range(step)])
y_pred = np.array([x[i][1] for i in range(step)])
z_pred = f(x_pred, y_pred)
x = np.arange(-10, 11, 2)
y = np.arange(-10, 11, 2)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
show3D(X, Y, Z, x_pred, y_pred, z_pred)
"""ケース3"""
# 初期点を設定してアルゴリズムを実行
x0 = np.array([30, 30])
x, f_history = steepest(x0, l, step, f, nabla_f)
print(x)
# グラフを描画
x_pred = np.array([x[i][0] for i in range(step)])
y_pred = np.array([x[i][1] for i in range(step)])
z_pred = f(x_pred, y_pred)
x = np.arange(-10, 11, 2)
y = np.arange(-10, 11, 2)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
show3D(X, Y, Z, x_pred, y_pred, z_pred)
```
