```python
"""最尤推定"""

import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

# ガウス分布
def gauss(mu, var, x):
    return np.exp(-(x-mu)**2 / 2*var*var) / np.sqrt(2*np.pi*var*var)

# 指数分布
def exponential(lam, x):
    return lam*np.exp(-lam*x) * (x>=0)
```

```python
# サンプルするデータ数，母平均，母分散を与えて最尤推定を行い，推定分布と真の分布の描画まで行う
def MostLikelihoodinNormal(n, mu, var):
    N = n # データ数
    s = np.random.normal(loc=mu, scale=np.sqrt(var), size=n) # データをサンプリング
    
    # 平均，分散の最尤推定量を計算
    e_mu = s.mean()
    e_var = ((s - e_mu)**2).mean()
    
    x = np.linspace(-5, 5, n)
    y_true = gauss(mu, var, x)
    y_esti = gauss(e_mu, e_var, x)
    
    plt.figure(figsize=(8,7))
    plt.plot(x, y_true, color='r', label='true')
    plt.plot(x, y_esti, color='b', label='estimated')
    plt.legend()
    plt.show()
```

```python
MostLikelihoodinNormal(1000, 0, 1)
```

```python
# サンプルするデータ数，母平均，母分散を与えて最尤推定を行い，推定分布と真の分布の描画まで行う
def MostLikelihoodinExp(n, lam):
    N = n # データ数
    s = np.random.exponential(scale=lam, size=n) # データをサンプリング
    
    # 指数分布のパラメータの最尤推定量を計算
    e_lam = s.mean()
    
    x = np.linspace(0, 100, n)
    y_true = exponential(1/lam, x)
    y_esti = exponential(1/e_lam, x)
    
    plt.figure(figsize=(8,7))
    plt.plot(x, y_true, color='r', label='true')
    plt.plot(x, y_esti, color='b', label='estimated')
    plt.legend()
    plt.show()
```

```python
MostLikelihoodinExp(100, 10)
```
