```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sympy import *
from sympy.plotting import plot_parametric, plot,plot_implicit
from sympy import Symbol, cos, sin, sqrt
```

```python
from pylab import rcParams
rcParams['font.family'] = 'Times New Roman' #全体のフォントを設定
rcParams["mathtext.fontset"] = "stix" 
rcParams["font.size"] = 15
rcParams['axes.linewidth'] = 1.0# 軸の線幅edge linewidth。囲みの太さ
rcParams['axes.grid'] = True
rcParams['figure.figsize'] = 6, 6
```

```python
t = Symbol('t')
p = plot_parametric(\
                (t, -t-1, (t, -2, 3)), \
                (t*(t-1), -abs(t), (t, -2, 3)), xlabel='$u$', ylabel='$t$', legend=True)
```

```python
p[0].line_color="red"
p[1].line_color="blue"
p[0].label="$t=-u-1$"
p[1].label="$(~x(x-1),~-|x|~)$"
p.show()
```

```python
t = Symbol('t')
p = plot_parametric((t*(t-1), -abs(t), (t, -1, 2)), xlabel='$u$', ylabel='$t$', line_color="blue")
```

```python
t = Symbol('t')
p = plot_parametric(\
                (-0.25, t, (t, -0.5, 1.5)), \
                (t*(t-1), -abs(t), (t, -1, 2)), \
                (t*(t-1), -abs(t), (t, 0.5, 2)), \
                    xlabel='$u$', ylabel='$t$')
p[0].line_color="red"
p[1].line_color="blue"
p[2].line_color="red"
p.show()
```

```python
t = Symbol('t')
p = plot_parametric(\
                (-0.25, t, (t, -0.5, 1.5)), \
                (t*(t-1), -abs(t), (t, -1, 2)), \
                (t*(t-1), -abs(t), (t, 0.5, 2)), \
                (t, -t-1, (t, -1, 1)), \
                    xlabel='$u$', ylabel='$t$', xlim=(-1, 2), ylim=(-2, 1.5))
p[0].line_color="red"
p[1].line_color="blue"
p[2].line_color="red"
p[3].line_color="green"
p.show()
```
